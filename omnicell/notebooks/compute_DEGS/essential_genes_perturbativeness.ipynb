```python

sc.pp.normalize_total(adata, target_sum=10_000)
sc.pp.log1p(adata)

```

```python
pert = 'CEBPZ'
cell = 'k562'
```

```python
adata_pert = adata[(adata.obs[dd.cell_key] == cell) & (adata.obs[dd.pert_key] == pert)]
adata_ctrl = adata[(adata.obs[dd.cell_key] == cell) & (adata.obs[dd.pert_key] == dd.control)]

print(f"# Control cells: {adata_ctrl.shape}")
print(f"# Perturbed cells: {adata_pert.shape}")
```

```python
degs = get_DEGs(adata_ctrl, adata_pert)
```

```python
pdefault=0.05

significant_degs = degs[degs['pvals_adj'] < pdefault]
```
