```python
import scanpy as sc
from omnicell.evaluation.utils import get_DEGs
from omnicell.data.catalogue import Catalogue

```

```python
DATA_PATH = '/om/group/abugoot/Projects/Omnicell_datasets/essential_gene_knockouts_raw/essential_gene_knockouts_raw.h5ad'



with open(DATA_PATH, 'rb') as f:
    adata = sc.read_h5ad(f)


adata
```

```python

sc.pp.normalize_total(adata, target_sum=10_000)
sc.pp.log1p(adata)

catalogue = Catalogue('configs/catalogue')
dd = catalogue.get_dataset_details('essential_gene_knockouts_raw')
```

```python
pert = 'CEBPZ'
cell = 'k562'
```

```python
adata_pert = adata[(adata.obs[dd.cell_key] == cell) & (adata.obs[dd.pert_key] == pert)]
adata_ctrl = adata[(adata.obs[dd.cell_key] == cell) & (adata.obs[dd.pert_key] == dd.control)]

print(f"# Control cells: {adata_ctrl.shape}")
print(f"# Perturbed cells: {adata_pert.shape}")
```

```python
degs = get_DEGs(adata_ctrl, adata_pert)
```

```python
pdefault=0.05

significant_degs = degs[degs['pvals_adj'] < pdefault]
```
