```python
import h5py
import torch
```

```python
file_path = "human_pt5.h5"
output_path = "gene_emb_prottrans5.pt"

# Open the h5 file
with h5py.File(file_path, "r") as h5_file:
    # Extract gene names and embeddings
    gene_names = list(h5_file.keys())
    repr_list = [torch.tensor(h5_file[gene_name][:]) for gene_name in gene_names]

# Stack embeddings into a single tensor
repr_tensor = torch.stack(repr_list)

# Create the desired dictionary format
data = {
    "gene_names": gene_names,
    "repr": repr_tensor
}
```
