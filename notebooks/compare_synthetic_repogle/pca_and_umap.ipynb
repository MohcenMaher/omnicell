```python
import matplotlib.pyplot as plt 
import scanpy as sc
```

```python
adata_true_pert = sc.read('/orcd/data/omarabu/001/Omnicell_datasets/satija_IFN_adjusted_truepert.h5ad')
adata_synthetic_pert = sc.read('/orcd/data/omarabu/001/Omnicell_datasets/satija_IFN_adjusted_pert.h5ad')
adata_true_ctrl = sc.read('/orcd/data/omarabu/001/Omnicell_datasets/satija_IFN_adjusted_tructrl.h5ad')
```

```python
adata_true_pert.obs["source"] = "true"
adata_synthetic_pert.obs["source"] = "synthetic"
adata_true_ctrl.obs["source"] = "control"
```

```python
PERT = "H1F0"
PERT_KEY = "gene"
adata_true_pert_sub = adata_true_pert[adata_true_pert.obs[PERT_KEY] == PERT]
adata_synthetic_pert_sub = adata_synthetic_pert[adata_synthetic_pert.obs[PERT_KEY] == PERT]
```

```python

adata_true_pert_sub.obs
```

```python
import scanpy as sc
import matplotlib.pyplot as plt

# Concatenate the datasets
adata_combined = adata_true_pert_sub.concatenate(adata_synthetic_pert_sub).concatenate(adata_true_ctrl)

# Preprocessing
sc.pp.normalize_total(adata_combined)
sc.pp.log1p(adata_combined)
sc.pp.highly_variable_genes(adata_combined, n_top_genes=2000)
adata_combined = adata_combined[:, adata_combined.var.highly_variable]
sc.pp.scale(adata_combined)
sc.pp.pca(adata_combined, n_comps=30)
sc.pp.neighbors(adata_combined)
sc.tl.umap(adata_combined)

# Create a figure with two subplots


```

```python
# UMAP plot
sc.pl.umap(adata_combined,
           color='source',
           title='UMAP: True vs Synthetic Perturbations H1F0')

adata_combined_synt_and_true = adata_combined[adata_combined.obs["source"].isin(["true", "synthetic"])]

# UMAP plot
sc.pl.umap(adata_combined_synt_and_true,
           color='source',
           title='UMAP: True vs Synthetic Perturbations H1F0') 

adata_combined_true_and_control = adata_combined[adata_combined.obs["source"].isin(["true", "control"])]
sc.pl.umap(adata_combined_true_and_control,
           color='source',
           title='UMAP: True vs Control H1F0')
```
